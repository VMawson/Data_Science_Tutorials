```python
# %load imports.py
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
diamonds=sns.load_dataset('diamonds')
print(diamonds.shape)


import random
def sys_sample(data, step):
    index=np.arange((random.randint(0,10)), len(data), step=step)
    sample=data.iloc[index]
    return sample
sample= sys_sample(diamonds,100)
```

## countplot

```python
%matplotlib inline
sns.countplot(x='cut', hue='color', data=sample, palette = 'hls')
```

## stripplot
representing categorical data on a scatter plot

```python
# jitter is the deviation from the true value so we can set this to false
sns.catplot(x="cut", y="price", data=sample, jitter=False)
```

```python
sns.catplot(x="cut", y="price", data=sample, jitter=False, hue='color')
```

## swarm plot

```python
sns.catplot(x="cut", y="price", data=sample, kind='swarm', s=2)
```

## violinplot
combines boxplot and kernel density estimation. quartile values are displayed inside the violin.

```python
sns.catplot(x="cut", y="price", data=sample, kind='violin')
# can also use the 'hue' feature for this plot. 
```

## boxplot

```python
sns.catplot(x="cut", y="price", data=sample, kind='bar')
```

## pointplot
displays the estimate valye abd confidence interval for the same hue category. helps to identify how the relationship is changing in a particular hue category.


```python
sns.catplot(x="cut", y="price", data=sample, kind='point', hue='color')
```

## boxplot
used to show quartile values of a distribution along with end values. Outliers are also plotted.


```python
sns.catplot(x="cut", y="price", data=sample, kind='box')
```
